## визуализация количественных данных

Для начала нарисуем простую кривую: $y=x^2$.
Рассчитываем векторы <b>x</b> и <b>y</b>, передаем их первыми двумя параметрами в функцию plot(), которая по умолчанию рисует график в виде линии.

````python
lens = [3, 8, 12, 14, 20]
# x=[i/10 for i in range(200)]
x=[i/10 for i in lens]
y=[i*i for i in x]

plt.plot(x, y)

plt.show()
````

Поиграем с цветом и формой элементов графика. Пусть это будет не кривая, а набор <font color="red">красных</font> (r) кружочков (o). (В скобках написаны буквы, которые надо передать в качестве строки; r означает "red", а "o" - это кружочек.)

````python
x=[i/20 for i in range(40)]
y=[i*i for i in x]

plt.plot(x, y,  'co')

plt.show()

# 'b'	blue
# 'g'	green
# 'r'	red
# 'c'	cyan
# 'm'	magenta
# 'y'	yellow
# 'k'	black
# 'w'	white
````

Или <font color="green"> зеленых</font> (g) треугольников углом вверх (^).

````python
x=[i/20 for i in range(0,40)]
y=[i*i for i in x]

plt.plot(x, y, 'm*')

plt.show()
````

А то и вовсе нарисуем несколько графиков разными цветами и формами точек (здесь два примера: один использует третий параметр, который сочетает в себе и цвет, и вид штриха; второй передает их разными параметрами - color и marker). 

Вот <a href="https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot">здесь</a> описаны все варианты задания цветов, точек и формы штриха.
Заодно добавим сетку - plt.grid().

````python
x=[i/20 for i in range(0,40)]
y=[i*i for i in x]

plt.plot(x, y, 'b--')
plt.plot(x, y[::-1], color='#6332a8', marker='^') # зеркальное отображение: шаг минус один - обратное направление
plt.plot(x, x, "g*" )

plt.grid() # задает сетку на графике

plt.show()

# коды цветов: https://www.google.com/search?q=color+picker
````

А еще добавим легенду. Для этого сперва сохраним наши графики в отдельные переменные plot1 и plot2.
После этого вызовем функцию plt.legend, в которую передается график (раз мы хотим его показать на легенде) и подпись к нему. loc = 'best' означает, что мы просим библиотеку найти лучшее с ее точки зрения размещение легенды.
Всё о легенде <a href="https://matplotlib.org/api/_as_gen/matplotlib.pyplot.legend.html#matplotlib.pyplot.legend">здесь</a>.

````python
x=[i/20 for i in range(0,40)]
y=[i*i for i in x]

plot1, =plt.plot(x, y, 'b--')     #перезапоминаем графики в переменные
plot2, =plt.plot(x, y[::-1], color='#9432a8', marker='^')

plt.grid()
plt.legend((plot1, plot2), ("График 1", "График 2"), loc = 'best') #задает легенду (описание графиков)

plt.show()
````

Всё то же самое можно было бы получить, добавив метку (label) к каждому графику, а потом попросив нарисовать легенду.<br>
Заодно окрасим легенду в другой цвет (<font color='#448844'>#448844</font>), увеличим размер шрифта (```fontsize=12```), добавим название графика ( ```plt.title()``` ) и название осей (```xlabel()``` и ```ylabel()```). В итоге легенда перебегает на другое место.

````python
x=[i/20 for i in range(0,40)]
y=[i*i for i in x]

plot1, =plt.plot(x, y, 'b--', label="График 1") # подписи графиков
plot2, =plt.plot(x, y[::-1], color='#FACE8D', marker='^', label="График 2") # название графилкв можно менять

plt.grid()
plt.title("Относительные чатсоты наших 2 графиокв", fontsize=20, color='#9adb40')
plt.xlabel("Ось X", fontsize=20) # оси соответвенно тоже можно называть по-разному
plt.ylabel("Ось Y", fontsize=20)
legend=plt.legend(loc = 'best', fontsize=12)
legend.get_frame().set_facecolor('#448844')


plt.show()
````

А теперь давайте попробуем рисовать при помощи функции scatter, которая работает несколько медленнее, но обладает большей функциональностью (см. <a href="https://matplotlib.org/api/_as_gen/matplotlib.pyplot.scatter.html#matplotlib.pyplot.scatter">здесь</a>).<br>


Так как scatter по определению - это набор точек, которые не должны соединяться между собой, а просто бросаются на плоскость, то отрисовать линию с его помощью не получится. <br>
Также в scatter нельзя передать сразу цвет и вид точки/штриха, поэтому используем параметры color и marker.


Параметр alpha можно применять почти везде. Он показывает степень прозрачности графика, легенды, ... 0 - полностью прозрачен, 1 - абсолютно непрозрачен. <br>
Помимо этого для каждой из точек можно задавать размер маркера (параметр s) и его цвет (параметр c).

````python
x=[i/20 for i in range(0,40)]
y=[i*i for i in x]
c=[i*i for i in x]  # маркер для цвета
# s=[i*i*100 for i in x] # маркер для размера
s=100

plt.scatter(x, y, c=c, s=s, alpha=1) # парамеир прозрачности от 0 до 1, c можно развернуть: [::-1]
plt.show()
````

Цвет графика можно задать при помощи цветовой карты. Для использования стандартных цветовых карт необходимо загрузить их из matplotlib: <br>
from matplotlib import cm <br>
А потом использовать ту, что больше нравится, в параметре cmap (а можно создать объект типа Colormap самостоятельно и передать туда же). В этом случае параметр <b>c</b> воспринимается как номер цвета в цветовой схеме, цветовая схема распределяется между минимальным и максимальным значениями. (И никто не мешал сделать то же самое с plot).

````python
from matplotlib import cm


x=[i/100 for i in range(0,100)]
y=[i*i for i in x]
c=[i*i for i in x]

plt.scatter(x, y, c=c, alpha=0.5, cmap=cm.jet) # можно изменять цветовую карту, например cm.summer, autumn, winter, https://matplotlib.org/3.1.0/tutorials/colors/colormaps.html

plt.show()
````
