```

#def preprocessed ():
data1 = data.lower()
data2 = data1.translate(str.maketrans('','',string.punctuation))
data3 = re.sub(r'[\d+]', '', data2)
data4 = re.sub(r'[\n+]', '', data3)
data5 = mystem.lemmatize(data4)
print(data5)   
```

```
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
nltk.download('punkt')


rustopwords = stopwords.words('russian')
rustopwords.extend(['который', 'это', 'свой', 'весь', 'твой', 'ваш', 'наш', ' '])
data6 = [word for word in data5 if word not in rustopwords]
print(sorted(data6))
```

```
from nltk.probability import FreqDist
fdist = FreqDist(data6)
fdist.most_common(10)
```

```
fdist.plot(30,cumulative=False)
```
