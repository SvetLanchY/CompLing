```
import requests as rq

from bs4 import BeautifulSoup
```

```
url = "https://codexlaw.ru/vse-stati"
page = rq.get(url)
page.status_code
```

```
page.text
```

```
import re
soup = BeautifulSoup(page.text, 'html.parser') 
print(soup.prettify())
```

```
for tag in soup.find_all('li'): # находит все заголовки тегов2
  print(re.sub(r'[\d,/+]', '', tag.text))
```
